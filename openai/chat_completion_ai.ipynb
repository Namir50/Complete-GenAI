```python
from openai import OpenAI
from dotenv import load_dotenv
import os
```

```python
load_dotenv()
```

```python
openai_key = os.getenv("OPENAI_API_KEY")
```

```python
client = OpenAI(api_key=openai_key)
```

```python
prompt = "What is the day today?"

response = client.chat.completions.create(
    model= "gpt-4.1-mini",
    messages= [
        {"role":"system","content":"You are a helpful assistant"},
        {"role":"user","content":prompt}
    ]
)

print(response)
print(response.choices[0].message.content)
```

usage of chat completion model

```python
prompt1 = "hello how are you?"
prompt2 = "My name is namir i am 23 years old"
prompt3 = "Tell me about me"

response = client.chat.completions.create(
    model = "gpt-4.1-mini",
    messages = [
        {'role':'system', 'content':'You are a helpful assistant'},
        {'role':'user','content':prompt1},
        {'role':'user','content':prompt2},
        {'role':'user','content':prompt3}
    ]
)
```

here it understands the question and answers based on previous question and response

tweeking parameters

```python
prompt = "What is the day today?"

response = client.chat.completions.create(
    model= "gpt-4.1-mini",
    messages= [
        {"role":"system","content":"You are a helpful assistant"},
        {"role":"user","content":prompt}
    ],
    max_tokens = 25
)

print(response)
print(response.choices[0].message.content)
```
