Messages are fundamental unit of context for models in langchain<break>
1. Role: indentifies the message type(eg: system, user)<break>
2. Content: respresents actual content of message(like textm audio, documents,etc)<break>
3. Metadata: optional fdields such as response info, message ids, token usage

```python
import os
from dotenv import load_dotenv

load_dotenv()
```

```python
from langchain.chat_models import init_chat_model

os.environ['GROQ_API_KEY'] = os.getenv('GROQ_API_KEY')

model = init_chat_model('groq:qwen/qwen3-32b')
```

```python
model.invoke("please tell me how ai works")
```

Text prompts: they are strings which are ideal for straight forward generation tasks where you dont need to retain conversation history

```python
model.invoke("hey there!")
```

Message prompts: here we can pass in a list of messsages to the model by providing list of message object
<break>
1. System message: tells the model how to behave and provide context for interactions<break>
2. Human message: represents user input<break>
3. AI message: responses generated by the model including text content, tool call, and meta data<break>
3. Tool message: represents the outputs of tool calls


Tool message

```python
from langchain.messages import HumanMessage,AIMessage,SystemMessage

messages = [
    SystemMessage("You are a professional poetry expert"),
    HumanMessage("Write a funny poetry on artificial intelligence")
]

response = model.invoke(messages)
print(response.content)
```

Detailed info to the LLM through system mesage

```python
from langchain.messages import SystemMessage, AIMessage, HumanMessage

systemmessage = SystemMessage("""You are a senior python developer withy expertise in web frameworks, 
                                 always provide cods examples and explain your reasoning, be concise""")

messages = [
    systemmessage,
    HumanMessage("How do i create a rest api?")
]

response = model.invoke(messages)
print(response.content)
```

Human message with metadata

```python
human_message = HumanMessage(
    content = 'Hello!',
    name = 'alice',
    id = "msg102"
)

response = model.invoke([human_message])
print(response.content)
```
