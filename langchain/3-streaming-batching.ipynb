most models can stream their output content while it is being generated, by displaying the output progressively, 
streaming significantly improves user experience. Calling stream() returns an iterator that yields output chunks as they are produced

```python
import os
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
os.environ['GOOGLE_API_KEY'] = os.getenv('GOOGLE_API_KEY')
os.environ['GROQ_API_KEY'] = os.getenv('GROQ_API_KEY')
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI

model = ChatGoogleGenerativeAI(model = 'gemini-2.5-flash-lite')
response = model.invoke("why is there so much heat in india during summers?")
response
```

```python
model.stream("why is there so much heat in india during summers?")
```

```python
for chunk in model.stream("why is there so much heat in india during summers?"):
    print(chunk.text, flush=True)
```

Batching a collection of independent requests to a model can significantly improve performance and reduce costs, as the processing can be done in parallel

```python
responses = model.batch([
    "why do parrots speak?",
    "how do planes fly",
    "what is quantum computing"
])

for response in responses:
    print(responses)
```

```python
responses = model.batch([
    "why do parrots speak?",
    "how do planes fly",
    "what is quantum computing"], 
    config={'max_concurrency': 10})

for response in responses:
    print(response.content)
```
