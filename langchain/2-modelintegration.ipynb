integrating gemini

```python
from langchain.chat_models import init_chat_model
model = init_chat_model("google_genai:gemini-2.5-flash-lite")
response = model.invoke("why do parrots talk?")
response
```

integrating gpt

```python
model = init_chat_model("gpt-4.1")
respone = model.invoke("best places in usa in terms of cleanliness")
response
```

another way for integrating gpt

```python
from langchain_openai import ChatOpenAI
model = ChatOpenAI(model = 'gpt-4.1')

response = model.invoke("how are you?")
response
```

another way of integrating gemini

```python
from langchain_google_genai import ChatGoogleGenerativeAI

model = ChatGoogleGenerativeAI(model = 'gemini-2.5-flash-lite')
response = model.invoke("why is there so much heat in india during summers?")
response
```
