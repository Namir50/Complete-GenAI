```python
import arxiv
import pandas as pd
```

```python
query = "ai OR artificial intelligence OR machine learning"

search = arxiv.Search(
    query = query,
    max_results = 1000,
    sort_by = arxiv.SortCriterion.SubmittedDate) #Sort papers by the date they were submitted to arXiv
```

```python
papers = []

for results in search.results():
  papers.append({
      'published':results.published,
      'title':results.title,
      'ABSTRACT':results.summary,
      'categories':results.categories,
  })
```

```python
df = pd.DataFrame(papers)

pd.set_option('display.max_colwidth',None)
df.head(10)
```
